```python
xs = np.linspace(5, 50)
ys = 3 * xs + norm(90, 10).rvs(50)
plt.plot(xs, ys)
```

```python
# plt.rcParams['font.family'] = 'sans-serif'
# plt.rcParams['font.sans-serif'] = 'Helvetica'

# PLOTLY EXPRESS

plt.rcParams['axes.edgecolor']= '#aaaaaa'
plt.rcParams['axes.linewidth']= 2
plt.rcParams['xtick.color']= 'red'
plt.rcParams['ytick.color']= 'red'

def basic_plot(xs, ys, figsize=(10,5), xlabel='X', ylabel='Y', title='Title', 
               alpha=0.5, marker='*', markersize=10, label='Label', color='#ff33cc', linewidth=1, linestyle='--',
               xlim=None, ylim=None, bgcolor='#eeeeee'):
    fig = plt.figure(figsize=figsize)
    ax = fig.add_subplot(1, 1, 1)
    ax.set_facecolor(bgcolor)
    ax.plot(xs, ys, alpha=alpha, marker=marker, markersize=markersize, label=label, color=color, linewidth=linewidth, linestyle=linestyle)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.set_title(title)
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
    ax.legend()
```

```python
# plt.style.available
# plt.style.use('ggplot')
basic_plot(xs, ys, figsize=(15,7))
```
